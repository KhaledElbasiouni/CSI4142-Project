# Project Deliverable: Data Staging

## Importing csv into dataframe

```python
import pandas as pd
```

```python
# Reading the csv files into dataframes
employment_df = pd.read_csv('data/employment-by-industry-monthly.csv')
```

```python
employment_df
```

## Filtering data

```python
filtered_employment_df = employment_df[(employment_df["Data type"] == "Trend-cycle") & (employment_df['REF_DATE'] >= '2000') & (employment_df['REF_DATE'] < '2021') & (employment_df["GEO"] == "Canada")]
```

## Dropping unrequired employment columns

```python
filtered_employment_df = filtered_employment_df.drop(columns=['DGUID', 'Statistics', 'Data type', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'SYMBOL', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS', 'SCALAR_FACTOR', 'UOM_ID', 'UOM', 'GEO'])
```

```python
print(filtered_employment_df.dtypes)
```

## Renaming columns

```python
filtered_employment_df.rename(columns={'REF_DATE': 'Date'}, inplace=True)
filtered_employment_df.rename(columns={'North American Industry Classification System (NAICS)': 'NAICS'}, inplace=True)
filtered_employment_df.rename(columns={'VALUE': 'People Employed'}, inplace=True)
```

```python
filtered_employment_df
```

## Changing data types

```python
filtered_employment_df['Date'] = pd.to_datetime(filtered_employment_df['Date']).dt.year
filtered_employment_df['Date'] = filtered_employment_df['Date'].astype('int64')
```

```python
filtered_employment_df.head()
```

```python
filtered_employment_df.dtypes
```

```python
filtered_employment_df[:-1]
```

```python
filtered_employment_df['People Employed'] = filtered_employment_df['People Employed'].apply(lambda x: x * 1000)
filtered_employment_df['People Employed'] = filtered_employment_df['People Employed'].astype('int64')
```

```python
filtered_employment_df
```

```python
filtered_employment_df = filtered_employment_df[(filtered_employment_df['NAICS'].str.contains(r'\['))]

sum_by_naics = filtered_employment_df.groupby(['Date','NAICS'])['People Employed'].sum()
sum_by_naics_df = sum_by_naics.reset_index()
sum_by_naics_df
```
