```python
import pandas as pd
```

```python
# Reading the csv files into dataframes
employment_df = pd.read_csv('data/employment-by-industry-monthly.csv')
```

## Dropping unrequired employment columns

```python
employment_df = employment_df.drop(columns=['DGUID', 'Statistics', 'Data type', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'SYMBOL', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS', 'SCALAR_FACTOR', 'UOM_ID'])
```

```python
print(employment_df.dtypes)
```

```python
employment_df.rename(columns={'REF_DATE': 'Date'}, inplace=True)
```

```python
employment_df['Date'] = employment_df['Date'].dt.to_period('M')
```

```python
filtered_employment_df = employment_df[employment_df['Date'] >= '2000-01']
```
